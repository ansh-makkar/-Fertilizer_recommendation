```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```python
df=pd.read_csv('Fertilizer_Prediction.csv')
df
```

```python
df.head()
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
y
```

```python
X['Crop'].unique()
```

```python
datamapping={
    
    'rice':1,
    'Coconut':2

}
```

```python
X['Crop']=X['Crop'].map(datamapping)
X.head()
```

```python
X['Soil'].unique()
```

```python
datamapping={
    

     'Clayey' : 1 ,
     'laterite' : 2 ,
     'silty clay' : 3 ,
     'sandy': 4 ,
     'coastal':5 ,
     'clay loam': 6,
     'alluvial'  :7

}
```

```python
X['Soil']=X['Soil'].map(datamapping)
```

```python
y.unique()
y
```

```python
datamapping={
    'DAP and MOP': 1, 
    'Good NPK':2, 
    'MOP':3,
    'Urea and DAP': 4,
    'Urea and MOP': 5,
    'Urea ':6,
    'DAP':7
    }
```

```python
y=y.map(datmapping)
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
```

```python
score=accuracy_score(y_test,y_predict)
score
```
